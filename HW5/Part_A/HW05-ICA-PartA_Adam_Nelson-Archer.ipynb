## Step-by-step explanation

The tutorial shows how to generate images that maximally activate individual filters in a ConvNet, so we can see what kinds of patterns each filter has learned to detect.

1. Load a pre-trained ResNet50V2 (ImageNet weights, no classification head) and pick an intermediate layer to inspect (`conv3_block4_out`). A sub-model is built that outputs just that layer's activations.

2. Define a loss function that takes an input image and a filter index, runs the image through the sub-model, and returns the mean activation of that filter (ignoring border pixels to avoid artifacts).

3. Implement a gradient ascent step. Using `tf.GradientTape`, compute the gradient of the loss with respect to the input image pixels, L2-normalize the gradients, and update the image in the direction that increases the filter's activation. This is the opposite of normal training -- instead of updating weights to minimize loss, we update the input image to maximize a filter's response.

4. To visualize a filter, start from a near-gray random image (small random values around zero) and run 30 gradient ascent steps with a learning rate of 10. After that, post-process the result: normalize to zero mean and unit variance, center-crop to remove border noise, and clip to valid pixel values (0-255).

5. Test on a single filter first (filter 0) to make sure the pipeline works, then loop over the first 64 filters and stitch the resulting images into an 8x8 grid.

6. The final grid shows different texture-like patterns for each filter -- things like stripes, dots, color blobs, checkerboards, etc. This is basically the "vocabulary" of visual features that the network uses to understand images. Filters in earlier layers tend to pick up simple edges and colors, while filters in deeper layers (like the one used here) respond to more complex patterns.
